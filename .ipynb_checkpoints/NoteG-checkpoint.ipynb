```python
import pygame
import numpy as np
import time
from IPython.display import HTML
import random
import threading
import types
import random
import traceback
from toggle_input import hide_toggle as hide
from collections import defaultdict 
def def_value(): 
    return None
stateManager = defaultdict(def_value) 

class GameThread(threading.Thread):
    def __init__(self,  *args, **kwargs):
        super(GameThread, self).__init__(*args, **kwargs)
        self._stop_event = threading.Event()

    def stop(self):
        self._stop_event.set()

    def stopped(self):
        return self._stop_event.is_set()
    
    
#Initial Config
class Game:
    def __init__(self,display_width,display_height,window_color,title):
        self.size = (display_width,display_height)
        self.title = title
        self.running = True
        self.scene = list()
        self.window_color=window_color
        self.event_handlers = dict()
        self.event_handlers[(('type',pygame.QUIT),)] = self.on_quit
        self.event_handlers[(('type',pygame.KEYDOWN), ('key',pygame.K_q))] = self.on_quit
        self.event_handlers[(('type',pygame.KEYDOWN), ('key',pygame.K_ESCAPE))] = self.on_quit
        self.flipdelay=16
        self.tickcounter=0
        
    def render(self):
        self.disp.fill((0,0,0))
        for obj in self.scene:
            try:
                obj.render(self.disp)
            except Exception:
                traceback.print_exc()
                self.scene.remove(obj)
                print("Rendering Error: Object "+str(obj)+" removed from the scene")
        # unlike .update() just updates the screen as a whole          
        pygame.display.flip()
    def update(self):
        dt=pygame.time.get_ticks()- self.tickcounter
        for obj in self.scene:
            try:
                obj.update(dt)
            except Exception:
                traceback.print_exc()
                self.scene.remove(obj)
                print("Update Error: Object "+str(obj)+" removed from the scene")
        self.tickcounter=pygame.time.get_ticks()
        pygame.time.delay(self.flipdelay)
    
    def on_quit(self, event):
        self.running = False
    
    def process_events(self):
        for event in pygame.event.get():
            dire = dir(event)
            for eh in self.event_handlers.keys():
                callit=True
                for (attrname,attrvalue) in eh:
                    if (not attrname in dire) or (event.__getattribute__(attrname)!=attrvalue):
                        callit=False
                        break
                if callit:
                    self.event_handlers[eh](event)
        
    def display(self):
        pygame.init()
        self.disp=pygame.display.set_mode(self.size, pygame.HWSURFACE | pygame.DOUBLEBUF)
        pygame.display.set_caption(self.title)
        self.clock=pygame.time.Clock()
        self.tickcounter=pygame.time.get_ticks()
        while( self.running ):
            try:
                self.render()
                self.process_events()
                self.update()
            except Exception:
                traceback.print_exc()
                pygame.time.delay(10000)
        pygame.quit()

def startGame(display_width=500,display_height=500,window_color=(0,0,0),title='NoteG Prototype'):
    global game,engine,scene
    game = Game(display_width,display_height,window_color,title)
    engine = GameThread(target = game.display)
    engine.start()

def getSize():
    return pygame.display.get_surface().get_size()
    
hide()
```

```python
class SpriteSheet(object):
    """ Class used to grab images out of a sprite sheet. """
    # This points to our sprite sheet image
    sprite_sheet = None

    def __init__(self, file_name):
        """ Constructor. Pass in the file name of the sprite sheet. """

        # Load the sprite sheet.
        self.sprite_sheet = pygame.image.load(file_name).convert()


    def get_image(self, x, y, width, height):
        """ Grab a single image out of a larger spritesheet
            Pass in the x, y location of the sprite
            and the width and height of the sprite. """

        # Create a new blank image
        image = pygame.Surface([width, height]).convert()

        # Copy the sprite from the large sheet onto the smaller image
        image.blit(self.sprite_sheet, (0, 0), (x, y, width, height))

        # Return the image
        return image
    
# for simple texture
class Tiles:
    def __init__(self,tile):
        self.tile = pygame.image.load(tile).convert()
        
    def update(self, dt):
        return
    
    def render(self, disp):
        for y in range(0,500,32):
            for x in range(0,500,32):
                disp.blit(self.tile, (x,y))

# for complex tile maps
class Tile:
    def __init__(self,grass,floor_one,wall_bottom,wall_right,wall_top,wall_left):
        self.grass=grass
        self.floor_one=floor_one
        self.wall_bottom=wall_bottom
        self.wall_right=wall_right
        self.wall_top=wall_top
        self.wall_left=wall_left

    def update(self,dt):
        return
    def render(self, disp):
        for y, row in enumerate(tile_map):
            for x, terrain_num in enumerate(row):
                if terrain_num == 0:
                    None

                elif terrain_num == 1:
                    tile = self.floor_one
                    rect = (x * 64, y * 64)

                elif terrain_num == 2:
                    tile = self.wall_bottom
                    rect = (x * 64, y * 64)

                elif terrain_num == 3:
                    tile = self.wall_right
                    rect = (x * 64, y * 64)

                elif terrain_num == 4:
                    tile = self.wall_top
                    rect = (x * 64, y * 64)

                elif terrain_num == 5:
                    tile = self.wall_left
                    rect = (x * 64, y * 64)

                disp.blit(tile, rect)
hide()
```

```python
class Character(pygame.sprite.Sprite):
    def __init__(self,char, x, y):
        pygame.sprite.Sprite.__init__(self)
        self.img=pygame.image.load(char)
        self.anim = dict()
        self.stat = dict()
        self.cycle_index = 0
        self.cycle_tick = 0
        seq = list()
        seq.append(80) # ticks per frame
        fin_img = ''
        for i in range(8):
            fin_img = self.img.subsurface((64+i*64,0,64,64))
            seq.append(fin_img)
        self.anim["up"] = seq
        self.stat["up"] = self.img.subsurface((0,0,64,64))
        width = fin_img.get_width()
        height = fin_img.get_height()
        self.rect = pygame.Rect(x, y, width, height)
        seq = list()
        seq.append(80) # ticks per frame
        for i in range(8):
            seq.append(self.img.subsurface((64+i*64,128,64,64)))
        self.anim["down"] = seq
        self.stat["down"] = self.img.subsurface((0,128,64,64))

        seq = list()
        seq.append(80) # ticks per frame
        for i in range(8):
            seq.append(self.img.subsurface((64+i*64,64,64,64)))
        self.anim["left"] = seq
        self.stat["left"] = self.img.subsurface((0,64,64,64))

        seq = list()
        seq.append(80) # ticks per frame
        for i in range(8):
            seq.append(self.img.subsurface((64+i*64,192,64,64)))
        self.anim["right"] = seq
        self.stat["right"] = self.img.subsurface((0,192,64,64))

        self.current_anim = "up"
        self.current_frames = self.anim[self.current_anim]
        self.pos = [x,y]
        self.speedx = 0
        self.speedy = 0

        self.last_update = pygame.time.get_ticks()
        self.frame = 0

        self.shoot_delay = 250
        self.last_shot = pygame.time.get_ticks()

    def update(self, dt):
        self.speedx = 0
        self.speedy = 0
        keystate = pygame.key.get_pressed()
        if keystate[pygame.K_LEFT]:
            self.speedx=-1
            self.current_anim="left"
        elif keystate[pygame.K_RIGHT]:
            self.speedx=1
            self.current_anim="right"
        if keystate[pygame.K_UP]:
            self.speedy=-1
            self.current_anim="up"
        if keystate[pygame.K_DOWN]:
            self.speedy=1
            self.current_anim="down"
        if keystate[pygame.K_RIGHT] and keystate[pygame.K_UP]:
            self.current_anim = "up"

        if keystate[pygame.K_LEFT] and keystate[pygame.K_UP]:
            self.current_anim = "up"

        if keystate[pygame.K_DOWN] and keystate[pygame.K_RIGHT]:
            self.current_anim = "down"

        if keystate[pygame.K_DOWN] and keystate[pygame.K_LEFT]:
            self.current_anim = "down"
#         if keystate[pygame.K_SPACE]:
#             self.shoot()
        ca = self.anim[self.current_anim]
        self.cycle_tick = (self.cycle_tick + dt) % ((len(ca)-1)*ca[0])
        self.cycle_index = int(self.cycle_tick/ca[0])
        dirs={"up":(0,-0.5),"down":(0,0.5), "right":(1,0), "left":(-1,0)}
        dxdy = dirs[self.current_anim]
        self.pos[0]+= self.speedx
        self.pos[1]+= self.speedy
#         self.rect.y += self.speedy
#         self.rect.x += self.speedx
        pass
        
    def render(self, display):
        ca = self.anim[self.current_anim]
        if self.speedx!=0 or self.speedy != 0:
            display.blit(ca[1+self.cycle_index], (self.pos))
        else:
            display.blit(self.stat[self.current_anim], (self.pos))
    def check_wall(self):
        wall_collisions = pygame.sprite.groupcollide(enemy_sprites, player_bullets, False, False)
        for collisions in wall_collisions:
            if tile.solid_state == True:
                self.speedx = 0
                self.speedy = 0
    
hide()
```

```python
# Sample Object class > Copy and modify as per required shape

# pygame.draw.rect
# pygame.draw.polygon
# pygame.draw.circle
# pygame.draw.ellipse
# pygame.draw.arc
# pygame.draw.line
# pygame.draw.lines
# pygame.draw.aaline
# pygame.draw.aalines

class Trinket(pygame.sprite.Sprite):
    def __init__(self,shape="circle",speed=4, image=None, center=None, **kwargs):
        pygame.sprite.Sprite.__init__(self)
        if not center==None:
            self.x = center[0]
            self.y = center[1]
        else:
#             self.x = random.randint(0,getSize()[0])
#             self.y = random.randint(-40,-20)
            self.x = game.player.pos[0]
            self.y = game.player.pos[1]
        self.dirx=1
        self.img=None
        if image!=None:
            if type(image) == str:
                self.img=pygame.image.load(image)
            else:
                self.img=image
            self.rect = self.img.get_rect()
        else:
            self.rect = kwargs.get('rect', (self.x,self.y,20,20))
        self.shape=shape
        self.speed=speed
        self.kwargs = kwargs
        
    def render(self, display):
        if not self.img:
            if self.shape == "circle":
                pygame.draw.circle(display, **self.kwargs, center=(self.x, self.y))
            else:
                # not checked for all shapes
                func = getattr(pygame.draw, shape)
                func(self.kwargs)
        else:
            display.blit(self.img,(self.x,self.y))        
    
    def update(self,dt):
        self.check_bullets()
        if self.x>500:
            self.dirx=-1
        if self.x<0:
            self.dirx=1
        self.x+=self.dirx*self.speed
        
    def check_bullets(self):
        hits = pygame.sprite.groupcollide(all_sprites, player_bullets, True, True)
        for hit in hits:
            self.kill()
hide()
```

```python
class parse100(pygame.sprite.Sprite):
    def __init__(self,char,t=0):
        pygame.sprite.Sprite.__init__(self)
        self.img=pygame.image.load(char)
        self.img.set_colorkey(DARK_GREEN)
        self.sprites = list()
        fin_img = ''
        for i in range(0,10):
            for j in range(0,10):
                if t==0:
                    fin_img = self.img.subsurface((4+i*60,4+i*66,58,64))
                elif t==1:
                    fin_img = self.img.subsurface((8+i*60,8+i*71,61,63))
                elif t==2:
                    fin_img = self.img.subsurface((4+i*124,4+i*104,122,102))
                self.sprites.append(fin_img)
    def random_obj(self):
        return random.choice(self.sprites)
```
